## Operadores

```r
# teste
result <- a * b + c / a - c^2
result
```

## Vetores | Função c()

Vetores é um conjunto de elementos da mesma natureza, a função c() é chamada de combinar, e é muito utilizada para criar vetores.

```r
# criando vetor numérico
numeros <- c(1,5,3,7,1,5)
'Vetor de números:'
numeros

# criando sequência de números
sequencia <- c(5:0)
'Sequência em ordem decrescente:'
sequencia

# vetor de strings
letras <- c('João','Pedro','Ana','Zack')
'Vetor de letras:'
letras

# vetor misto
mistura <- c('abacate',2,'banana',3)
"Vetor misto"
mistura
```

<span class="mark">OBSERVAÇÕES:</span><br> >> Para criar uma sequência, é utilizado : (dois pontos), seguido da sequência de números desejados<br> >> Ao criar um vetor misto, note que os números passam a ser strings

## Condições 

<strong>Sintaxe</strong><br>
> if ( condição for verdadeira ) {<br>
----comando para executar<br>
} else {<br>
----comando para executar<br>
}

```r
# exemplo
if (a > c) {
    print("Olá R")
} else {
    print("É menor")
}
```

<strong>Sintaxe</strong>

> ifelse (testa, se verdadeiro, se falso)


```r
ifelse (c == a, "é igual", "não é igual")
```

## Loops

### loop for

> for ( variável in lista ) { instrução }

```r
lista <- c(1,4,32,76,85)

# utilizando FOR
for (i in lista) {
    print(i)
}
```

### loop while

> while ( condição ) { instrução }

```r
x <- 0
while (x < length(lista)) {
    x <- x + 1
    print(x)
}
```

## Definir funções

<strong>Sintaxe</strong>

> variavel <- function ( arg1, arg2... ) { estrutura da função }

```r
somar <- function (x, y) {
    sum <- x + y
    return (sum)
}
somar(2, 6)
```
